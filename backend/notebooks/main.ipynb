```python
from faster_whisper import WhisperModel

model = WhisperModel(
    "small",
    device="cpu",
    compute_type="int8",
    download_root=Path.cwd().parent / "models",
)
```

```python
result = model.transcribe("capture/audio_vocals.mp3", beam_size=5)
```

```python
transcripts, info = result
for segment in transcripts:
    print(f"[{segment.start:.2f}s - {segment.end:.2f}s] {segment.text}")
```
