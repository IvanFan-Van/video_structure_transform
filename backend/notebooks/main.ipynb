```python
import os
import sys

from dotenv import load_dotenv, find_dotenv
from pathlib import Path

load_dotenv(find_dotenv(), override=True)

sys.path.append(str(Path(os.getcwd()).parent))
```

```python
VIDEO_FILENAME = "抖音2026528-645764.mp4"
VIDEO_PATH = Path.cwd().parent / "tests" / "videos" / VIDEO_FILENAME
```

```python
from openai import OpenAI

client = OpenAI(
    # 火山引擎的 API Key，对应你截图中的 api-key
    api_key=os.getenv("API_KEY", ""),
    # 火山引擎的官方 OpenAI 兼容接入地址
    base_url=os.getenv("BASE_URL", ""),
)
```

```python
from ffmpeg import FFmpeg

ffmpeg = (
    FFmpeg()
    .input(VIDEO_PATH)
    .output("output.mp4", {"codec:v": "libx264"}, vf="scale=1280:720", crf=24)
)

ffmpeg.execute()
```

```python
file = client.files.create(
    file=open("output.mp4", "rb"),
    purpose="user_data",
    extra_body={
        "preprocess_config": {
            "video": {"fps": 0.3},
        }
    },
)
file
```

```python
SYSTEM_PROMPT = """"""
VIDEO_ANALYZE_PROMPT = """"""
VIDEO_GENERATE_PROMPT = """"""

try:
    # 发起对话请求
    completion = client.chat.completions.create(
        # 注意：这里必须填入你的 Endpoint ID (EP)
        model=os.getenv("MODEL", ""),
        messages=[
            {"role": "system", "content": SYSTEM_PROMPT},
            {"role": "user", "content": VIDEO_ANALYZE_PROMPT},
        ],
        # 传输视频
        temperature=0.7,
    )

    # 打印返回结果
    print(completion.choices[0].message.content)

except Exception as e:
    print(f"调用失败，错误信息: {e}")
```
