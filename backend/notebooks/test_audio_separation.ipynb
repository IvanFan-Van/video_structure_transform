```python
from pathlib import Path

video_path = Path("../tests/videos/抖音2026528-272271.mp4")
output_dir = Path("./capture")
output_dir.mkdir(exist_ok=True)
```

```python
import ffmpeg
from audio_separator.separator import Separator

audio_path = output_dir / "audio1.wav"
ffmpeg.input(str(video_path)).output(str(audio_path)).run(overwrite_output=True)
separator = Separator(output_dir=str(output_dir), output_format="mp3")

separator.load_model("UVR-MDX-NET-Inst_HQ_3.onnx")
output_names = {
    "Vocals": "vocals_output",
    "Instrumental": "instrumental_output",
}
output_files = separator.separate(str(audio_path), output_names)
```
